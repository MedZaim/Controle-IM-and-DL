```python
import pandas as pd

# Load the original metadata with semicolon separator
original_csv = 'HAM5000/HAM500_metadata.csv'
df = pd.read_csv(original_csv, sep=';')

# Create binary label: 'mel' → 1, others → 0
df['binary_label'] = df['label'].apply(lambda x: 1 if str(x).strip() == 'mel' else 0)

# Optional: verify class distribution
print("Original label distribution:")
print(df['label'].value_counts())
print("\nBinary label distribution:")
print(df['binary_label'].value_counts())
```

```python
import pandas as pd
import os

# Paths
input_csv = 'etape1/HAM5000/HAM500_metadata.csv'
output_csv = 'etape1/HAM5000/HAM500_metadata_binary.csv'

# Load original CSV (semicolon-separated)
df = pd.read_csv(input_csv, sep=';')

# Create binary label column
df['binary_label'] = df['label'].apply(lambda x: 1 if str(x).strip() == 'mel' else 0)

# Save new CSV with standard comma separator
df.to_csv(output_csv, index=False)

# Verify
print(" New CSV created successfully.")
print(f"Path: {os.path.abspath(output_csv)}")
print("\nFirst 3 rows:")
print(df[['image_id', 'label', 'binary_label']].head(3))
```
